```python
from CRUD_Python_Module import AnimalShelter
```

```python
shelter = AnimalShelter('aacuser', '***')
print('Connection successful')
```

```python
new_animal = {
    "animal_type": "Dog",
    "breed": "Labrador Retriever Mix",
    "name": "TestDog",
    "age_upon_outcome": "2 years",
    "outcome_type": "Adoption",
    "sex_upon_outcome": "Neutered Male",
    "color": "Black"
}
result = shelter.create(new_animal)
print("Create result:", result)
```

```python
result = shelter.read({"name": "TestDog"})
print("Read result:")
for doc in result:
    print(doc)
```

```python
count = shelter.update(
    {"name": "TestDog"},
    {"name": "UpdatedTestDog", "age_upon_outcome": "3 years"}
)
print("Documents modified:", count)

result = shelter.read({"name": "UpdatedTestDog"})
print("\nVerification read:")
for doc in result:
    print(doc)
```

```python
count = shelter.delete({"name": "UpdatedTestDog"})
print("Documents deleted:", count)

result = shelter.read({"name": "UpdatedTestDog"})
print("Verification read (should be empty):", result)
```
